```python
import os
from dotenv import load_dotenv 
```

```python
load_dotenv()  # Load environment variables from .env file
```

```python
print(os.environ.get("MESH_API_URL"))  # Print the value of OPENAI_API_KEY
```

```python
from langchain.agents import create_agent
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
    api_key=os.environ["MESH_API_KEY"],
    base_url=os.environ["MESH_API_URL"].rstrip("/"),
    model=os.environ["MESH_MODEL"],
    temperature=0.2,
)
```

```python
def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"


agent = create_agent(
    model=llm,
    tools=[get_weather],
    system_prompt="You are a helpful assistant",
)

result = agent.invoke(
    {"messages": [{"role": "user", "content": "tell me 5lines about pune?"}]}
)
print(result["messages"][-1].content_blocks)
```
